```python
import requests
import json
import pandas as pd

base_url = "https://api.openalex.org"
info = "/authors"
API = "5NnEdiLWysIU50NJxbfgDB"
complete_url = base_url + info

# Use a list to collect author data
authors_list = []

# loop through all authors and store their information in the list
while complete_url: 
    response = requests.get(complete_url, params={"mailto": API})
    print("status code:", response.status_code)
    data = response.json()
    print("response:", data)
    for author in data["results"]:
        authors_list.append({
            "id": author["id"],
            "display_name": author["display_name"],
            "works_api_url": author["work_api_url"],
            "h_index": author["h_index"],
            "works_count": author["works_count"],
            "country_code": author["country_code"]
        })
    complete_url = data["meta"]["next"]

# Create DataFrame from the list
authors_df = pd.DataFrame(authors_list)

# save the dataframe to a CSV file
authors_df.to_csv("authors.csv", index=False)
```

```python
info2 = "/works"
complete_url2 = base_url + info2
works_list = []

while complete_url2:
    response = requests.get(complete_url2, params={"mailto": API})
    print("status code:", response.status_code)
    data = response.json()
    print("response:", data)
    for work in data["results"]:
        works_list.append({
            "id": work["id"],
            "publication_year": work["publication_year"],
            "cited_by_count": work["cited_by_count"],
            "author_ids": work["author_ids"],
            "title": work["title"],
            "abstract_inverted_index": work["abstract_inverted_index"]
        })
    complete_url2 = data["meta"]["next"]
works_df = pd.DataFrame(works_list)
works_df.to_csv("works.csv", index=False)
```
